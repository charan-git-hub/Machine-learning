## Machine Learning lifecycle

### 1. Study the Problem
- Understand the business problem and define objectives for the model.
- Collaborate with stakeholders to clarify goals and success criteria.
- Identify whether the problem requires classification, regression, clustering, or other ML techniques.

### 2. Data Collection
- Gather data from various sources such as databases, APIs, or manual surveys.
- Ensure that the data is relevant, accurate, and complete.
- Document the data collection process for reproducibility.

### 3. Data Preparation

 **a. Data Cleaning**
- Handle missing values by imputation or removal.
- Remove duplicates and correct data inconsistencies.
- Identify and treat outliers using statistical methods.

 **b. Data Transformation**
- Scale numerical features (e.g., normalization or standardization).
- Encode categorical variables using techniques like one-hot or label encoding.
- Combine or split features if needed (e.g., extracting year from a date).

**c. Exploratory Data Analysis (EDA) and Feature Engineering**
- Visualize data distributions, correlations, and trends using tools like Matplotlib and Seaborn.
- Create new features or transform existing ones to improve model performance.

 **d. Dataset Splitting**
- Divide the dataset into training, validation, and testing sets (e.g., 70%-20%-10%).

### 4. Model Selection
- Choose an appropriate algorithm based on the problem type and dataset characteristics.
- Evaluate the pros and cons of algorithms like Decision Trees, SVMs, Neural Networks, etc.
- Consider computational efficiency and scalability.

### 5. Model Building and Training
- Define the model architecture (for deep learning) or choose a traditional ML model.
- Specify hyperparameters like learning rate, number of layers, or tree depth.
- Train the model on the preprocessed training data.
- Use techniques like cross-validation to improve generalizability.

### 6. Model Evaluation
- Test the model using the validation or test set.
- Use evaluation metrics suitable for the task:
  - Classification: Accuracy, Precision, Recall, F1-score, ROC-AUC.
  - Regression: Mean Squared Error (MSE), Mean Absolute Error (MAE), R-squared.
- Compare results with baseline models or benchmarks.

### 7. Model Tuning
- Optimize hyperparameters using grid search, random search, or Bayesian optimization.
- Regularize the model to prevent overfitting.
- Use advanced techniques like ensemble learning or fine-tuning for better performance.

### 8. Deployment
- Deploy the trained model as a web service or API for real-time predictions.
- Ensure the deployment environment supports scalability and reliability.
- Create user-friendly interfaces for interacting with the model.

### 9. Monitoring and Maintenance
- Continuously monitor the model's performance in production.
- Retrain the model periodically with new data to handle concept drift.
- Address any issues like model degradation or ethical concerns.
- Document changes and updates for transparency.


## Types of Machine Learning

## 1. Supervised Machine Learning

- Supervised learning is the most common type of machine learning.
  
- In this approach, the model is trained on a labeled dataset, where input-output parameters are already mapped.
  
- Enables machines to predict outputs based on provided training.
  
- In supervised learning, sample labeled data are provided to the machine learning system for training, and the system then predicts the output based on the training data.

- The system uses labeled data to build a model that understands the datasets and learns about each one. After the training and processing are done, we test the model with sample data to see if it can accurately predict the output.





- Supervised learning can be grouped further in two categories of algorithms:

+ **Classification**
  
+ **Regression**

### Categories of Supervised Learning:

 **1. Regression**

 Regression algorithms are used if there is a relationship between the input variable and the output variable. It is used for the prediction of continuous variables, such as Weather forecasting, Market Trends, etc. Below are some popular Regression algorithms which come under supervised learning:

- Linear Regression
- Regression Trees
- Non-Linear Regression
- Bayesian Linear Regression
- Polynomial Regressionession

**2. Classification**

Classification algorithms are used when the output variable is categorical, which means there are two classes such as Yes-No, Male-Female, True-false, etc.

## 2. Unsupervised Machine Learning

- The training data is unlabeled, and the algorithm acts on data without any supervision.
  
- The goal is to restructure input data into new features or groups based on patterns.

- Unsupervised learning is a learning method in which a machine learns without any supervision.
- 
The training is provided to the machine with the set of data that has not been labeled, classified, or categorized, and the algorithm needs to act on that data without any supervisio

- . The goal of unsupervised learning is to restructure the input data into new features or a group of objects with similar patterns.

### Types of Unsupervised Learning Algorithm:

The unsupervised learning algorithm can be further categorized into two types of problems:



**a. Clustering:**

- Groups objects into clusters based on similarities or differences.

-  Example: Grouping customers by the products they purchase.

**b. Association:**

- Identifies typical relations between variables in large datasets.
  
- Example: Market basket analysis for finding product associations.

## 3. Reinforcement Learning

- Feedback-based learning where the agent gets rewards for correct actions and penalties for wrong actions.

- The agent interacts with the environment to explore and learn.
  
- The goal is to maximize cumulative rewards.
  
- Example: Teaching a robot to navigate a maze or teaching an AI to play chess.
